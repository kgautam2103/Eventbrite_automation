```python
import pandas as pd
import requests
import json
from datetime import datetime
import time

#define global variables
auth_token = ''
prod_url = 'https://www.eventbriteapi.com/v3/'
prod_proxy_url= 'https://private-anon-a95bf0df64-eventbriteapiv3public.apiary-proxy.com/v3/'
mock_url='https://private-anon-86d996fd56-eventbriteapiv3public.apiary-mock.com/v3/'
org = 'organizations/'
venues= '/venues/'
events= 'events/'
copy= '/copy/'
publish= '/publish/'
current_ts = datetime.now().strftime('%Y%m%d%H%M%S')
output_sheet = current_ts+'_publish_output'
counter=1


authorization = 'Bearer ' + auth_token
auth_headers = {
    'Authorization': authorization,
    'Content-Type': 'application/json'
    }

wb = gc.open_by_url('https://docs.google.com/spreadsheets/d/1N0ZnO_6GsixL13sLahCda_8j8q6_bkos3W4yWS_qSGk/edit#gid=0')

# create and Open our new sheet and add some data.
sh = gc.create(output_sheet)
output_sheet = gc.open(output_sheet).sheet1

output_sheet.insert_row(['event_id','is_published', 'error'],1)

#input_sheet_name = input("enter the input sheet name: ")
input_data = wb.worksheet("inputdata")
inputdata = input_data.get_all_values()
inputdata_df = pd.DataFrame(inputdata, columns=['event_id'])
inputdata_df = inputdata_df.drop(inputdata_df.index[0])


for index, row in inputdata_df.iterrows():
  counter = counter+1
  is_published = False
  publish_error = ""
  cell_value_list = []
  event_id = str(row['event_id']).strip()
  publish_event_url = prod_url+events+event_id+publish
  time.sleep(60)
  publish_response = requests.post(publish_event_url,headers=auth_headers)
  publish_status = publish_response.status_code
  print("event_id "+event_id)
  print(publish_response.text)
  if publish_status == 200:
    is_published = True
  else:
    publish_error = publish_response.text

  cell_value_list = [event_id,is_published,publish_error]
  output_sheet.insert_row(cell_value_list,counter)
```
